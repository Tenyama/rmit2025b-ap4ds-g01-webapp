```python
import pandas as pd

file_path = '../data/assignment3_II.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
df = pd.DataFrame(data)
new_df = df[["Clothing ID", "Division Name", "Department Name", "Class Name", "Clothes Title", "Clothes Description"]]
```

```python
new_df.info()
```

```python
new_df.head(5)
```

```python
new_df = new_df.drop_duplicates()
```

```python
new_df.info()
```

```python
new_df = new_df.rename(columns={
    "Clothing ID": "id",
    "Division Name": "division",
    "Department Name": "department",
    "Class Name": "class",
    "Clothes Title": "title",
    "Clothes Description": "description"
})
```

```python
new_df = new_df.sort_values(by='id')
```

```python
# Does not make sense because each Clothing ID should only correspond to 1 item
# => Only keep 1 item from each duplicated id
new_df[new_df.duplicated(subset=["id"], keep=False)]
```

```python
# Drop duplicated and only keep the first one
new_df = new_df.drop_duplicates(subset=["id"], keep="first")

# Double check
new_df[new_df.duplicated(subset=["id"], keep=False)]
```
